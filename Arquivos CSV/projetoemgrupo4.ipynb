Projeto para gerar relatório dos preços da Gasolina e do Etanol dos meses Maio(5) e Junho(6) de 2023.

```python
# importa as bibliotecas
import pandas as pd
import numpy as np
```

```python
# Abrir arquivos dos dados dos precos da gasolina e etanol.
precos_gaso_et5 = pd.read_csv('precosgemes5.csv', sep=';')
precos_gaso_et6 = pd.read_csv('precosgemes6.csv', sep=';')
display(precos_gaso_et5)
display(precos_gaso_et6)
```

```python
# Quantidade de linhas e colunas da tabela do mes 5 e 6.
display(precos_gaso_et5.shape)
display(precos_gaso_et6.shape)
```

```python
# Concatenar as tabelas e juntas os indices
# pode ser esse tambem mas ele nao junta os indices keys=['mes5', 'mes6']
tabelas_preco_gasol_etan = pd.concat([precos_gaso_et5, precos_gaso_et6], ignore_index=True)
tabelas_preco_gasol_etan
```

```python
# Mostra a quantidade de linhas e colunas.
tabelas_preco_gasol_etan.shape
```

```python
# Mostra todas as colunas.
tabelas_preco_gasol_etan.columns
```

```python
# Deletando colunas usando o drop para fazer a limpeza adequada da tabela.
tabelas_preco_gasol_etan2 = tabelas_preco_gasol_etan.drop(columns=['Revenda','CNPJ da Revenda','Valor de Compra','Nome da Rua','Numero Rua','Complemento','Bairro','Cep','Unidade de Medida'])
```

```python
# Renomear nomes das colunas
tabelas_preco_gasol_etan3 = tabelas_preco_gasol_etan2.rename(columns=lambda x: x.replace('Regiao - Sigla', 'Sigla_regiao').replace('Estado - Sigla','Sigla_estado').replace('Data da Coleta','Data_coleta').replace('Valor de Venda','Valor_venda',))
tabelas_preco_gasol_etan3
```

```python
# Ver a quantidade de valores nulos em cada coluna.
valores_nulos = tabelas_preco_gasol_etan3.isna().sum()
valores_nulos
```

```python
# Substituir vírgulas por pontos e converter para float
tabelas_preco_gasol_etan3['Valor_venda'] = tabelas_preco_gasol_etan3['Valor_venda'].str.replace(',', '.').astype(float)
```

```python
# Agrupar os dados pelo nome do produto e calcular a soma das vendas total para cada produto
total_vendas_por_produto = tabelas_preco_gasol_etan3.groupby('Produto')['Valor_venda'].sum()
# Resetar o índice e renomear a coluna
total_vendas_por_produto = total_vendas_por_produto.reset_index()
total_vendas_por_produto.rename(columns={'Valor_venda': 'Total_vendas'}, inplace=True)
total_vendas_por_produto
```

```python
# preco Minimo
preco_minimo = tabelas_preco_gasol_etan3.groupby('Produto')['Valor_venda'].min()
# Resetar o índice e renomear a coluna
preco_minimo = preco_minimo.reset_index()
preco_minimo.rename(columns={'Valor_venda': 'Preco_minimo'}, inplace=True)
preco_minimo
```

```python
# preco Maximo
preco_maximo = tabelas_preco_gasol_etan3.groupby('Produto')['Valor_venda'].max()
# Resetar o índice e renomear a coluna
preco_maximo = preco_maximo.reset_index()
preco_maximo.rename(columns={'Valor_venda': 'Preco_maximo'}, inplace=True)
preco_maximo
```

```python
# Somar valor total das vendas dos 03 produtos: Etanol, Gasolina e Gasolina Aditivada.
soma_valordevendatotal = tabelas_preco_gasol_etan3['Valor_venda'].sum()
soma_valordevendatotal
```
